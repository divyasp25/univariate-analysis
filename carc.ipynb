```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load your dataset
# Replace 'your_dataset.csv' with the actual path to your dataset
df = pd.read_csv("C:\\Users\\Divya Shree\\Downloads\\creditcard (1).csv")

# 1. Filter the data based on some logic
# Example: Filtering rows where the 'Amount' is greater than 100
filtered_df = df[df['Amount'] > 100].copy()

# 2. Rename the columns if required
filtered_df.rename(columns={'V5': '5V'}, inplace=True)

# 3. Show the relationship between the variables using correlation
correlation_matrix = filtered_df.corr()

# a. Represent the association in visual form (using a heatmap)
plt.figure(figsize=(12, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt='.2f')
plt.title('Correlation Heatmap')
plt.show()

# 4. Count the number of rows
row_count = len(df)
print(f"Number of rows: {row_count}")

# 5. Detect outliers
# a. Represent the outliers in visual form (using a boxplot)
plt.figure(figsize=(12, 6))
sns.boxplot(data=filtered_df, orient="v")
plt.title('Boxplots for Attributes')
plt.xticks(rotation=90)
plt.show()

# b. Apply different methods to handle outliers (e.g., winsorization)
from scipy.stats.mstats import winsorize
for column in filtered_df.columns:
    if column != 'Class':  # Exclude 'Class' column
        filtered_df[column] = winsorize(filtered_df[column], limits=[0.05, 0.05])

# 6. Check if there is categorical data to convert in this dataset.
if 'Amount' in filtered_df.columns:
    filtered_df = pd.get_dummies(filtered_df, columns=['Amount'])

# 7. Drop irrelevant columns
# Example: Dropping 'irrelevant_column' if it exists
if 'Class' in filtered_df:
    filtered_df = filtered_df.drop(columns=['Class'])

# Now, you can use filtered_df for further analysis or save it to a new dataset.
print(filtered_df)
```
